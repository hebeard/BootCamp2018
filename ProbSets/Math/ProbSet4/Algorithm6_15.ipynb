```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg as la
```

```python
def f(x):
    return x**4 - 14*x**3 + 60*x**2 - 70*x
```

```python
def f_pr(x):
    return 4 * x**3 - 14 * 3 * x**2 + 60 * 2 * x - 70
```

```python
def problem6_15(x_0, x_1, ε, f_pr):
    MAX_IT = 1_000
    k = 0
    x_k = x_0
    x_kp1 = x_1
    X = np.linspace(-10,10,1_000)
    while k < MAX_IT:
        x_km1 = x_k
        x_k = x_kp1
        x_kp1 = x_k - f_pr(x_k) * (x_k - x_km1)/(f_pr(x_k) - f_pr(x_km1))
        k += 1
        plt.plot(X, f(X), "b--")
        plt.plot(x_k, f(x_k), "ko")
        plt.show()
        if la.norm(x_kp1 - x_k) < ε * la.norm(x_k):
            break
    return x_k
```

```python
problem6_15(0, 0.1, 1e-10, f_pr)
```
